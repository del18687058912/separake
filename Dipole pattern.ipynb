```python
d = 2      # meters
c = 343    # m/s
f = 3000   # Hertz
n_points = 1000

th = np.linspace(0, 360, n_points) * np.pi/180

fig = plt.figure(figsize=(10, 10))

for i, f in enumerate([1000, 2000, 3000, 4000]):
    fig.add_subplot(2, 2, i+1, polar=True)
    resp = np.abs(1 - np.exp(1j*d*f/c*np.sin(th)))
    plt.polar(th, resp)
```
